### Data exploration with ydata-profiling

```python
from ydata_profiling import ProfileReport
```

```python
profile = ProfileReport(randomSample, title='Education Attainment And Enrolment Around The World', explorative=True)
```

```python
profile.to_notebook_iframe()
```

### Summary

The dataset provides a detailed analysis of grades for different age groups (15-49, 20-29, 30-39, and 40-49) across genders (male and female) and their distribution over the years. Here are the key insights:

1. **Overall Trends:**
   - The grades for all age groups and genders are relatively high, with means generally above 0.7.
   - There is a noticeable trend where the younger age groups tend to have higher grades compared to the older age groups.

2. **Age Group Comparisons:**
   - **Ages 15-49:** This group shows the highest mean grades across all three categories (grade 1, grade 2, and grade 3). Grade 1 has the highest mean (0.8739), followed by grade 2 (0.8615), and grade 3 (0.8375).
   - **Ages 20-29:** This age group also shows high grades but slightly lower than the 15-49 age group. The mean grades are around 0.79 to 0.82.
   - **Ages 30-39:** The grades start to decline in this group with means ranging from 0.72 to 0.77.
   - **Ages 40-49:** This group has the lowest grades among all age groups, with means ranging from 0.66 to 0.71.

3. **Gender Comparisons:**
   - **Males:** Across all age groups, males tend to have slightly higher grades than females. For instance, in the 15-49 age group, the mean grade for males is 0.8946 (grade 1), while for females, it is 0.8535.
   - **Females:** Although females have slightly lower grades than males, their grades are still high, with means generally above 0.6 for all age groups.

4. **Distribution and Variability:**
   - The standard deviations are relatively consistent within each age group and gender, indicating similar variability in grades.
   - The minimum grades show significant variability, especially in the older age groups (e.g., Ages 40-49 with a minimum of around 0.03).

5. **Yearly Distribution:**
   - The year distribution indicates that the dataset spans from 1981 to 2020, with a mean year of approximately 2005. This suggests a broad temporal coverage, allowing for longitudinal analysis.

Overall, the dataset reveals that younger age groups and males tend to have higher grades, and there is a general decline in grades as age increases. The consistency in variability suggests stable grading patterns across different demographics.

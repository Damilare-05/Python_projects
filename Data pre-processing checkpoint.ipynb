### Data cleaning

#### Inspecting the dataset for potential missing values.

```python
print(df.isnull().sum())
```

**From above we can see the 'counter_number ' and 'reading_remarque' are the only columns that contain missing values**

#### Selecting strategy to handle missing values, and why I made that choice.

#### 'counter_number ' 

```python
print(f"Mean: {df['counter_number'].mean()}")
print(f"Median: {df['counter_number'].median()}")
print(f"Mode: {df['counter_number'].max()}")
```

**Since 'counter_number' seems to have a wide range of values (from 0 to 2.798115e+13), replacing missing values with the median might be a suitable approach. The median is less sensitive to outliers compared to the mean and can provide a representative central value.**

```python
df['counter_number'] = df['counter_number'].fillna(df['counter_number'].mean())
```

#### 'reading_remarque' 

```python
print(df['reading_remarque'].mean())
print(df['reading_remarque'].median())
print(df['reading_remarque'].mode())
```

**'reading_remarque' represents numerical data and I don't suspect heavy skewness or outliers based on my initial exploration, using the mean to replace missing values is a reasonable choice.**

```python
df['reading_remarque'] = df['reading_remarque'].fillna(df['reading_remarque'].mean())
```
